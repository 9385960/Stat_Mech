# The Van der Waals Potential

## The Van der Waals Equation for a real gas.

The equation of state for a real Van der Waals gas is given as follows:

$\left(p+\left(\frac{N}{V}\right)^2a\right)\left(V-Nb\right)=NkT$

```python
class VWGas:
	
	R = 8.31446261815324 #J/K/mol

	def __init__(self, a, b, T):
		self.a = a
		self.b = b
		self.T = T

	def get_pressure(self,V):
		c1 = self.R*self.T
		c2 = V-self.b
		c3 = (1/V)**2*self.a
		return (c1/c2) - c3
```

# Parameters for Real Gases

| Material | a ($Pa \cdot m^6 \cdot mol^{-2}$) | b ($10^{-3} \cdot m^3 \cdot mol^{-1}$) |
|----------|----------|----------|
| $H_2$    | 0.01945     | 0.022     |
| $H_2O$    | 0.056539     | 0.031     |
| $N_2$    | 0.13882     | 0.039     |
| $O_2$    | 0.13983     | 0.032     |
| $CO_2$    | 0.37186     | .043     |

```python
volumes = np.linspace(0.04e-3,1e-3,1000)

temps = [50, 100,200, 300, 374]
gases = []
for temp in temps:
	gases.append(VWGas(0.13882,0.039e-3,temp))

for gas in gases:
	plt.plot(volumes,gas.get_pressure(volumes), label=str(gas.T))

plt.title("P vs. V for 1 mol of $N_2$")
plt.xlabel("Volume ($m^3$)")
plt.ylabel("Pressure (Pa)")
plt.ylim(-300e5,500e5)
plt.legend()
```
